Q1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even(numbers):
  even_sum = 0
  for num in numbers:
    if num % 2 == 0:
      even_sum += num
  return even_sum

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_even(nums))
```

Q4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
  if num < 2:
    return False
  for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
      return False
  return True

for n in range(1, 201):
  if is_prime(n):
    print(n)
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class fibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms

        self.count = 0

        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration

        elif self.count == 0:
              self.count += 1
              return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = fibonacciIterator(10)

for num in fib:
    print(num, end=" ")
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(n):
  """Generator that yields powers of 2 up to exponent n ."""
  for i in range(n + 1):
    yield 2 ** i

for value in power_of_two(5):
  print(value)
```

Q7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
  """Generator that reads a file line by line and yields each line."""
  try:
    with open(filename, 'r') as f:
      for line in f:
        yield line.strip() # Using strip() to remove leading/trailing whitespace including newlines
  except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
  except Exception as e:
    print(f"An error occurred: {e}")

# Create a dummy file for testing
with open("example.txt", "w") as f:
  f.write("This is line 1.\n")
  f.write("This is line 2.\n")
  f.write("This is line 3.")

# Call the generator function
for line in read_file_line_by_line("example.txt"):
  print(line)
```

Q9.  Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("celsius: ", celsius)
print("fahrenheit: ", fahrenheit)
```

Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

```python
def remove_vowels(s):
  vowels = "aeiouAEIOU"

  result = "".join(filter(lambda ch: ch not in vowels, s))
  return result

text = "Hello World, Python  is amazing!"
print("original:", text)
print("without vowels:", remove_vowels(text))
```

 Write a Python program using lambda and map.
